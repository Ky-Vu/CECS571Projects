```python
##### Part 1 #####
from rdflib import Graph
g = Graph()
g.parse("filmgraph.ttl")

import pprint
for stmt in g:
  pprint.pprint(stmt)
```

```python
##### Part 2 #####

from rdflib.namespace import DC, DCTERMS, DOAP, FOAF, SKOS, OWL, RDF, VOID, XMLNS, RDFS, XSD, Namespace
from rdflib import URIRef, Literal

bruce = URIRef("http://dbpedia.org/resource/Bruce_Willis")
for s, p, o in g:
  if (g.predicate_objects(subject=bruce, unique=False)):
    print("\nPredicate: ", p, "\nObject: ", o, "\n")
```

```python
##### Part 3 #####
n = Namespace("http://example.org/")
dbo = Namespace("http://dbpedia.org/ontology/")
dbr = Namespace("http://dbpedia.org/resource/")
diehard = URIRef('http://example.org/DieHard')

m = Graph()
m.add((diehard, RDF.type, dbo.Film))
m.add((diehard, dbo.starring, dbr.Bruce_Willis))
```

```python
##### Part 4 #####
from rdflib.tools.rdf2dot import rdf2dot
from IPython.display import display, Image
import pydotplus, io

def visualize(g):
  stream = io.StringIO()
  rdf2dot(g, stream, opts = {display})
  dg = pydotplus.graph_from_dot_data(stream.getvalue())
  png = dg.create_png()
  display(Image(png))

visualize(m)
```

```python
##### Part 5 #####
m.add((dbr.Bruce_Willis, n.portrays, dbr.John_McClane))
```

```python
##### Part 6 #####
g.serialize(destination="g_graph.ttl")
m.serialize(destination="m_graph.ttl")
o = Graph()
o.parse("g_graph.ttl")
o.parse("m_graph.ttl")
```

```python
##### Part 7 #####

visualize(o)
```
